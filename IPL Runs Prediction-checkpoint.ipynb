# Import the necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

# Import the Dataset

```python
df=pd.read_excel(r'C:\Users\Amrit\Desktop\Data.xlsx')
```

```python
df.head()
```

# Data Cleaning

```python
## Check for null values
df.isnull().sum()
```

```python
# Encoding the names of the players
le=LabelEncoder()
df.PLAYER=le.fit_transform(df.PLAYER)
```

```python
df.head()
```

```python
# Renaming the column 2018_Runs 
df.rename(columns={'2018_Runs':'Prev_year_Runs'},inplace=True)
```

```python
# The HS column has elements as strings.Converting the strings to integers
for i in range(100):
    num_str=''
    if type(df.iloc[i,5])==str:
        for j in df.iloc[i,5]:
            if j.isdigit():
                num_str+=j
        df.iloc[i,5]=num_str
df.HS=df.HS.astype(int)
```

```python
# Splitting the dataset into independent and dependent features
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

# Exploratory Data Analysis and Feature Selection

```python
corr=X.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr,annot=True)
```

```python
# The 'Avg'column is obtained by the following formula
# df.Avg=df.2018_Runs/(df.Inns-df.NO)
#SO 'Avg'column is a redundant feature and needs to be removed
X.drop('Avg',axis=1,inplace=True)
```

```python
columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.90:
            if columns[j]:
                columns[j] = False
selected_columns = X.columns[columns]
```

```python
selected_columns  ## The relevant features as obtained from correlation matrix
```

```python
X.drop([columns for columns in X if columns not in selected_columns],axis=1,inplace=True)
```

```python
X
```

```python
sns.pairplot(df)
```

# Splitting the data into training and testing set

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
# The 'PLAYER' does not participate in the algorithm building
players_train=X_train['PLAYER']
player_test=X_test['PLAYER']
```

```python
X_train.drop('PLAYER',axis=1,inplace=True)
X_test.drop('PLAYER',axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

# Machine Learning Model -Supervised

    The dataset seems to have a significant degree of non linear relationship among features and target as obtained from the pair plot.Based on this inference some of the algorithms have beed tried.

# K-Nearest Neighbors

```python
knn=KNeighborsRegressor()
knn.fit(X_train_scaled,Y_train)
y_pred_knn=knn.predict(X_test_scaled)
```

```python
print('MSE :',mean_squared_error(Y_test,y_pred_knn),',','R2_score :', r2_score(Y_test,y_pred_knn))
```

# Decision Tree Regression

```python
dt=DecisionTreeRegressor()
dt.fit(X_train_scaled,Y_train)
y_pred_dt=dt.predict(X_test_scaled)
```

```python
print('MSE :',mean_squared_error(Y_test,y_pred_dt),',','R2_score :', r2_score(Y_test,y_pred_dt))
```

# Random Forest Regression

```python
rf=RandomForestRegressor()
rf.fit(X_train_scaled,Y_train)
y_pred_rf=rf.predict(X_test_scaled)
```

```python
print('MSE :',mean_squared_error(Y_test,y_pred_rf),',','R2_score :', r2_score(Y_test,y_pred_rf))
```

# Gradient Boosting Regression

```python
gbr=GradientBoostingRegressor()
gbr.fit(X_train_scaled,Y_train)
y_pred_gbr=gbr.predict(X_test_scaled)
```

```python
print('MSE :',mean_squared_error(Y_test,y_pred_gbr),',','R2_score :', r2_score(Y_test,y_pred_gbr))
```

# Hyperparameter Tuning of Random Forest Model

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.arange(50,110,10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(50, 110, num = 7)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               #'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               #'min_samples_leaf': min_samples_leaf,
               'bootstrap' : bootstrap}
```

```python
rf_random = GridSearchCV(estimator = rf, param_grid = param_grid, scoring='neg_mean_squared_error', cv = 5, verbose=2,  n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train_scaled,Y_train)
```

```python
rf_random.best_params_ # These are the optimum hyperparameters for the model.
```

```python
# Running the model with optimised hyperparameters
rf2=RandomForestRegressor(n_estimators=50,max_features='auto',min_samples_split=2,bootstrap=True,random_state=10)
rf2.fit(X_train_scaled,Y_train)
y_pred_test=rf2.predict(X_test_scaled)
y_pred_train=rf2.predict(X_train_scaled)
```

# Model Performance

```python
print('Testing Error :',mean_squared_error(Y_test,y_pred_test))
print('Training Error :',mean_squared_error(Y_train,y_pred_train))
```

```python
print('R2 Score :',r2_score(Y_test,y_pred_test))
```
